# CREATING NUMPY ARRAYS

```python
import numpy as np

#1-d array
a = np.array([1,2,3])
print(a)
```

```python
# 2-d array
b = np.array([[1,2,3], [4,5,6]])
print(b)
```

```python
# 3-d array
c = np.array([[[1,2], [3,4]], [[5,6], [7,8]]])
print(c)
```

```python
# creating numpy arrays with a specific data type
d = np.array([1,2,3], dtype=float)
print(d)

# creating array of bool type
e = np.array([1,2,3], dtype=bool)
print(e)
```

```python
# np.arange() - creates a numpy array with numbers within a given range
f = np.arange(1,9,2)
print(f)
```

```python
# np.reshape() -  reshapes a numpy array to a given number of rows and cols if possible
np.arange(1,11).reshape(5,2)
```

```python
# np.ones() and np.zeros()
print(np.zeros((3,2)))
print(np.ones((3,2)))
```

```python
# np.random() - generates a numpy array with random numbers between 0 and 1
np.random.random((3,2))
```

```python
# np.lisnpace() - creates an array within the given range with equidistant numbers
np.linspace(-10,10,10) # generates an array with 10 equidistant numbers between -10 and 10
```

```python
# np.identity() - creates an identity matrix
np.identity(3)
```

# ARRAY ATTRIBUTES


```python
# creating some demo arrays
a1 = np.arange(10)
a2 = np.arange(12, dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
# ndim - returns the dimensions of the array
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
# shape - returns shape of the array
print(a1.shape)
print(a2.shape)
print(a3.shape)
```

```python
# size - returns number of elemetns in the array
print(a1.size)
print(a2.size)
print(a3.size)
```

```python
# itemsize - returns the size occupied by each item of the array in the memory
print(a1.itemsize)
print(a2.itemsize)
print(a3.itemsize)
```

```python
# dtype - returns the datatype of the array
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

# Changing datatype of an array

```python
a1.astype(np.int32)
```

# ARRAY OPERATIONS

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
```

```python
#  SCALAR OPERATIONS
print("a1+a2", a1+a2)
print("-------------------------------")
print("a1-a2", a1-a2)
print("-------------------------------")
print("a1*a2", a1*a2)
print("-------------------------------")
print("a1**2", a1**2)
```

```python
# RELATIONAL OPERATORS
print("a1>a2", a1>a2)
```
