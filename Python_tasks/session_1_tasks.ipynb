3### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
x1 = float(input("Enter x-coordinate of the first point: "))
y1 = float(input("Enter y-coordinate of the first point: "))
x2 = float(input("Enter x-coordinate of the second point: "))
y2 = float(input("Enter y-coordinate of the second point: "))

def euclidean_distance(x1,y1,x2,y2):
  return round(((x2-x2)**2 + (y2-y1)**2) ** 0.5, 2)

euclidean_distance(x1,y1,x2,y2)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
# Write your code here
principal = float(input("Enter principal amount: "))
interest_rate = float(input("Enter interest rate: "))
time_period = float(input("Enter time period: "))

def calculate_simple_interest(principal, interest_rate, time):
  return round(float((principal*interest_rate*time)/100), 2)

calculate_simple_interest(principal, interest_rate, time_period)
````


### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





### Logic behind solving this
let number of dogs be d and number of chickens be c.
- each animal has one head : c + d= total_heads (let this be th)
- a chicken has 2 legs and a dog has 4 legs: 2c + 4d = total_legs (let this be tl)
- now solve the 2 equations.
- multiply the first equation by 2 -> 2c + 2d = 2*tl
- now equation2 - 2*equation 1 = 2c + 4d - 2c - 2d = tl - 2*th
- this gives us 2d = tl - 2*th
- d = (tl - 2*th)/2 or number of dogs = (total_legs - 2*total_heads)/2
- and from the first equation, c = total_heads - d
- so now we have solutions for both c and d

````python
# Write your code here
num_heads = int(input("Enter number of heads: "))
num_legs = int(input("Enter number of legs: "))

num_dogs = (num_legs-(2*num_heads))/2
num_chickens = num_heads - num_dogs

print(f"dogs : {num_dogs}, chickens: {num_chickens}")
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = int(input("Enter a value for n: "))
sum_ = 0

for i in range(1,n+1):
  sum_ += i

print(f"sum of first {n} natural numbers is: {sum_}")
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
num1 = int(input("Enter the first term of the arithmetic series: "))
num2 = int(input("Enter the second term of the arithmetic series: "))
n = int(input("Enter nth term to be found: "))

common_difference = num2-num1
nth_term = num1 + (n-1)*common_difference
print(f"{n} term of this series is: {nth_term}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm
