# INTRODUCTION
What is Pandas?
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool built on top of Python programming language.

What is a pandas series?
Pandas series - like a column of a table - holds 1-d data of any data type

# Creating pandas series


```python
# creating a series from a list
lst = ["India", "USA", "Dubai"]
a = pd.Series(lst)
a
```

```python
# creating series from integers
subjects = ["Math", "Science"]
marks = [89,95]
series = pd.Series(marks, index=subjects, name="marks")
series
```

# Attributes of a series
