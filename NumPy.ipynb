# Why does numpy exist?
Numpy exists because
1. numpy arrays are faster than built-in data types in python like python lists.
2. Numpy arrays occupy lesser memory space as compared to python lists.
3. Numpy arrays are convenient. to add numbers in 2 python lists, we'd have to use loops but using numpy arrays, we can just add the 2 arrays.

```python
# demonstrating that operations performed using numpy arrays are faster as comprared to those performed usng python lists.

import numpy as np
import time

print ("using Python lists")

l1 = [i for i in range(10000000)]  # initiating a list with 10 million numbers
l2 = [i for i in range(10000000, 20000000)]
l3 = []

start = time.time()
for i in range(len(l1)):
  l3.append(l1[i] + l2[i])
print(f"Time taken by Python list to add 10000000 items in 2 lists: ", time.time()- start)


# performing the same operation using Numpy arrays
print("Using Numpy arrays")
a1 = np.arange(10000000)
a2 = np.arange(10000000, 20000000)
start2 = time.time()
a3 = a1 + a2    # itemwise addition in numpy arrays so no loops required
print(f"Time taken by Numpy arrays to add 10 million items: ", time.time()- start2)
```

```python
# numpy arrays occupy lesser memory space as compared to python lists
import sys
import numpy as np

lst_ = [i for i in range(10000000)]
print(f"memory occupied by python list containing 10 million numbers: ", sys.getsizeof(lst_))

arr = np.arange(10000000, dtype=np.int16)
print(f"memory occupied by python list containing 10 million numbers: ", sys.getsizeof(arr))
```

# Broadcasting in numpy
- Broadcasting is how numpy deals with arrays with different shapes during arithmetic operations.
- The smaller array is broadcast across the larger array so they have the same size.

image.png

# Broadcasting rules

1. Make the 2 arrays have the same number of dimensions. If 2 arrays have different dimensions, add a dimension of size 1 to the head of the array with smaller dimensions.

2. Make each dimension of the 2 arrays have the same size. If size of the dimensions are differnet in 2 arrays, then dimensions with size 1 are streched to match the size of the arrays.

However, if there is a dimension whose size is not 1 in either of the arrays, then broadcasting can't be done and an error is raised.

```python
# adding arrasy iwth same size
arr1 = np.arange(1,7).reshape(2,3)
arr2=  np.arange(6,12).reshape(2,3)
arr1+arr2
```

```python
# but adding arrays of 2 differnet sizes is also possible
arr3 = np.arange(1,7).reshape(2,3)
arr4 = np.arange(6,9).reshape(1,3)

arr3+arr4
```

# Some useful numpy functions


```python
# np.sort() - returns a sorted copy of an array

a = np.random.randint(0,100,15)
print("ascending order:")
print(np.sort(a))
print("----------------------------------------")
print("Descending order")
print(np.sort(a)[::-1])
```

```python
# sorting in 2d array
b = np.random.randint(0,100,24).reshape(6,4)
print("Row-wise sorting")
print(np.sort(b))
print('------------------')
print("Column-wise sorting")
print(np.sort(b, axis=0))
```

```python
# append a 2-d array
np.append(b, np.random.random((b.shape[0], 1)), axis=1)  # appends a column at the end. to append a row, set axis=0
```

```python
# np.concatenate() - concatenates 2 arrays based on a given axis
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print("Row-wise concatenation")
np.concatenate((c,d), axis=0)  # pass a tuple which has the arrays that need to be concatenated
```
